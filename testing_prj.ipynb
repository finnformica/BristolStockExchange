## Testing PRJ
- Developed PRJ as an extension of PRDE with the JADE algorithm
- Implemented into BSE inheriting from PRZI
- Test that the selection and mutation works as intended

```python
from BSE import market_session
```

```python
# configure traders
def traders_config(k, f=0.5, c=0.1, p=0.2, no_traders=5):
    params = {
        'k': k, 
        'F': f,
        'c': c,
        'p': p,
        's_min': -1.0,
        's_max': +1.0,
        'wait_time': 7200
    }
    
    sellers_spec = [('PRJ', no_traders, params)]
    buyers_spec = sellers_spec
    
    return {'sellers': sellers_spec, 'buyers': buyers_spec}
```

```python
# confiure the supply and demand
sup_range = (100, 200)
dem_range = (100, 200)

stepmode = 'jittered'

# configure order schedules
no_days = 2
# no_days = 1/24/60 *3
start_time = 0
end_time = int(no_days*60*60*24)
# end_time = 30 * 60 * 60
supply_schedule = [{'from': start_time, 'to': end_time, 'ranges': [sup_range], 'stepmode': stepmode}]
demand_schedule = [{'from': start_time, 'to': end_time, 'ranges': [dem_range], 'stepmode': stepmode}]

order_interval = 10
timemode = 'drip-jitter'
order_sched = {
    'sup': supply_schedule,
    'dem': demand_schedule,
    'interval': order_interval,
    'timemode': timemode
}

traders_spec = traders_config(4)
```

```python
trial_id = 'output/testing_prj/trial_1'
tdump = open(trial_id + '_avg_balance.csv','w')
dump_all = True
verbose = False

# run market session
market_session(trial_id, start_time, end_time, traders_spec, order_sched, tdump, dump_all, verbose)

tdump.close()
```
